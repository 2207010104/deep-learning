```python
import os

os.makedirs(os.path.join('..', 'data'), exist_ok=True)
data_file = os.path.join('..', 'data', 'test.csv')
with open(data_file, 'w',encoding='utf8') as f:
    f.write('num,name,gender,age\n')  # 列名
    f.write('99,"小红","女",10\n')  # 每行表示一个数据样本
    f.write('98,"小蓝","男",NA\n')
    f.write('NA,"小黄","女",12\n')
    f.write('96,"小黑","男",NA\n')
    f.write('95,"小白","女",10\n')
    f.write('94,NA,"男",NA\n')
    
```

```python
import pandas as pd

data = pd.read_csv(data_file)
data
```

```python
null_counts = data.isnull().sum()
null_counts
```

```python
most_null_column = null_counts[null_counts == null_counts.max()].index[0]
most_null_column
```

```python
data = data.drop(most_null_column,axis=1)
```

```python
data
```

```python
import torch

inputs = data.iloc[:,0:3]
inputs = inputs.fillna(inputs.select_dtypes(include='number').mean())
inputs
```

```python
inputs = pd.get_dummies(inputs,dummy_na=True)
inputs
```
